```python
%%writefile app.py 
#####################################################
#Importamos librerias
import streamlit as st
from PIL import Image
import plotly.express as px
import pandas as pd
from streamlit_option_menu import option_menu
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

#####################################################
#Dise침o de pesta침a
img_icon = Image.open("icon.png") #cargar la imagen 
st.set_page_config(
    page_title="Dashboard Florencia",
    page_icon= img_icon,
    layout="wide",
)

#####################################################
#Definimos la instancia
@st.cache_resource

######################################################
#Creamos la funci칩n de carga de datos
def load_data():
   #Lectura del archivo csv
   df = pd.read_csv("Florencia_limpio.csv")
    
   #Selecciono las columnas tipo numericas del dataframe
   numeric_df = df.select_dtypes(['float','int'])  #Devuelve Columnas
   numeric_cols= numeric_df.columns                #Devuelve lista de Columnas

   #Selecciono las columnas tipo texto del dataframe
   text_df = df.select_dtypes(['object'])  #Devuelve Columnas
   text_cols= text_df.columns              #Devuelve lista de Columnas
   
   #Selecciono algunas columnas categoricas de valores para desplegar en diferentes cuadros
   categorical_column_sex= df['room_type']
   #Obtengo los valores unicos de la columna categ칩rica seleccionada
   unique_categories_sex= categorical_column_sex.unique()

   return df, numeric_cols, text_cols, unique_categories_sex, numeric_df

######################################################
#Cargo los datos obtenidos de la funci칩n "load_data"
df, numeric_cols, text_cols, unique_categories_sex, numeric_df = load_data()

######################################################
#CREACI칍N DEL DASHBOARD
#Generamos las p치ginas que utilizaremos en el dise침o
#Widget 1: Selectbox
#Menu desplegable de opciones de las p치ginas seleccionadas
View= st.selectbox(label= "View", options= ["INICIO","EXTRACCI칍N DE CARACTER칈STICAS 游늭","REGRESI칍N LINEAL SIMPLE 游늳","REGRESI칍N LINEAL MULTIPLE 游늵", "REGRESI칍N LOG칈STICA 游댝"])
# CONTENIDO DE LA VISTA 1
if View == "INICIO":
    st.title("Florencia, Italia")
    st.write("Este dashboard presenta un Modelado explicativo usando un an치lisis univariado de las variables categ칩ricas m치s significativas y un Modelado predictivo usando un an치lisis al aplicar regresi칩n lineal simple 游늳, regresi칩n lineal multiple y regresi칩n logistica, esto usando datos de Airbnb acerca de la ciudad de Florencia, Italia.")
    img, title = st.columns([1, 7])
    left, right = st.columns([3, 1])
    img.image("airbnb.png", width=80)
    title.header("Acerca de Airbnb")
    left.subheader("쯈u칠 es?")
    left.write("Airbnb es una plataforma digital que conecta a personas que desean alquilar su propiedad con viajeros que buscan alojamiento temporal. Fundada en 2008, Airbnb ha transformado la industria del hospedaje, ofreciendo alternativas m치s flexibles y personalizadas que los hoteles tradicionales.A trav칠s de su modelo de econom칤a colaborativa, permite que anfitriones publiquen espacios disponibles y que hu칠spedes puedan reservarlos de forma segura, utilizando filtros como precio, ubicaci칩n, tipo de propiedad, calificaciones, y m치s.")
    right.image("airbnbFoto.jpg", width=400)
    st.subheader("Estadisticas generales")
    st.markdown("""
                                    - Ingresos promedio por anfitri칩n: 30,000 euros al a침o
                                    - Tasa de ocupaci칩n media: 74%
                                    - Reservas anuales proemdio por propiedas: 270 noches
                                    - Mes con mayores reservar: mayo
                                    """)
    img, title = st.columns([1, 7])
    left, right = st.columns([1, 3])
    img.image("bandera.png", width=80)
    title.header("Acerca de Florencia")
    right.subheader("쯇or qu칠 elegir Florencia?")

    right.write("Florencia es una de las joyas culturales de Italia, considerada cuna del Renacimiento y hogar de algunas de las obras de arte y arquitectura m치s influyentes del mundo. Su atm칩sfera elegante, sus calles adoquinadas llenas de historia y su importante legado art칤stico la hacen un destino imperdible tanto para turistas como para estudiosos del arte y la cultura.")

    right.write("En el contexto de Airbnb, Florencia representa un mercado tur칤stico consolidado y muy competitivo, con una amplia gama de alojamientos que van desde apartamentos hist칩ricos hasta modernos espacios dise침ados para viajeros. Barrios como Santa Croce, Santo Spirito o el centro hist칩rico son zonas clave con alta demanda, ideales para analizar tendencias de hospedaje urbano.")
    left.image("FlorenciaDuomo.jpeg", width=440)
    right.image("FlorenciaPan.jpg", width=700)
    left.image("ribollita.jpeg", width=440)
    left.image("PonteVecchio.jpeg", width=440)
    st.subheader("Datos relevantes culturales")
    st.markdown("""
                                    - Nombres conocidos: La cuna del Renacimiento, la ciudad del arte, y la joya de la Toscana.
                                    -  Poblaci칩n: Aproximadamente 366,000 habitantes.
                                    -  Principales lugares tur칤sticos:
                                        - Catedral de Santa Maria del Fiore (Duomo)
                                        - Galleria degli Uffizi
                                        - Ponte Vecchio
                                        - Palazzo Pitti
                                        - Piazza della signoria
                                    -  Gastronom칤a ic칩nica:
                                        - Bistecca alla fiorentina
                                        - Ribollita
                                        - Pappa al pomodoro
                                    """)

###############################################################################
# CONTENIDO DE LA VISTA 2
elif View == "EXTRACCI칍N DE CARACTER칈STICAS 游늭":
    #Generamos los encabezados para el dashboard
    st.title("Extracci칩n de Caracter칤sticas")
    st.header("Panel Principal")
    st.write(df)
    st.write(df.columns)
    #######################
    ####Generamoos sidebar
    st.sidebar.title("Extracci칩n de Caracter칤sticas")
    st.sidebar.header("Sidebar")

    ###############################
    #widget 2: Checkbok
##condicional para que aparezca el dataframe

###############################################################################
# CONTENIDO DE LA VISTA 3
elif View == "REGRESI칍N LINEAL SIMPLE 游늳":
#Generamos los encabezados para el dashboard
    st.title("REGRESI칍N LINEAL SIMPLE")
    st.title("Modelado Predictivo")
    
    # Sidebar para selecci칩n de variables
    st.sidebar.title("Configuraci칩n del Modelo")
    numeric_cols = [col for col in numeric_cols if df[col].nunique() > 1]  # Filtra variables constantes
    
    # Selectores de variables
    feature = st.sidebar.selectbox("Variable Independiente (X)", numeric_cols)
    target = st.sidebar.selectbox("Variable Dependiente (y)", numeric_cols)
    
    # Divisi칩n de datos
    X = df[[feature]]
    y = df[target]
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # Modelado
    model = LinearRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    # M칠tricas
    st.subheader("Resultados del Modelo")
    st.write(f"Coeficiente: {model.coef_[0]:.4f}")
    st.write(f"Intercepto: {model.intercept_:.4f}")
    st.write(f"R: {model.score(X_test, y_test):.4f}")
    
    # Gr치fico
    fig = px.scatter(x=X_test[feature], y=y_test, opacity=0.7, 
                     labels={'x': feature, 'y': target},
                     title=f"Regresi칩n: {target} ~ {feature}")
    fig.add_scatter(x=X_test[feature], y=y_pred, mode='lines', name='L칤nea de Regresi칩n')
    st.plotly_chart(fig, use_container_width=True)
    
    
    
# CONTENIDO DE LA VISTA 4
elif View == "REGRESI칍N LINEAL MULTIPLE 游늵":
#Generamos los encabezados para el dashboard
    st.title("REGRESI칍N LINEAL MULTIPLE")
    st.title("Modelado Predictivo")
    st.header("Regresi칩n Lineal M칰ltiple")
    
    st.sidebar.title("Configuraci칩n del Modelo")
    selected_features = st.sidebar.multiselect("Variables Independientes (X)", numeric_cols)
    target = st.sidebar.selectbox("Variable Dependiente (y)", numeric_cols)
    
    if len(selected_features) >= 2:
        X = df[selected_features]
        y = df[target]
        
        # Estandarizaci칩n
        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)
        
        X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
        
        model = LinearRegression()
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        
        st.subheader("Resultados del Modelo")
        st.write("Coeficientes:", dict(zip(selected_features, model.coef_)))
        st.write(f"Intercepto: {model.intercept_:.4f}")
        st.write(f"R: {model.score(X_test, y_test):.4f}")
        
        # Gr치fico de residuos
        residuals = y_test - y_pred
        fig = px.scatter(x=y_pred, y=residuals,
                         labels={'x': 'Predicciones', 'y': 'Residuos'},
                         title="An치lisis de Residuos")
        fig.add_hline(y=0, line_dash="dash")
        st.plotly_chart(fig, use_container_width=True)
    else:
        st.warning("Selecciona al menos 2 variables predictoras")

# CONTENIDO DE LA VISTA 4
elif View == "REGRESI칍N LOG칈STICA 游댝":
#Generamos los encabezados para el dashboard
    st.title("TITANIC")
    st.header("Panel Principal")
    st.subheader("Bar Plot")
    
    #Menus desplegables de opciones de la variables seleccionadas
    Variable_cat= st.sidebar.selectbox(label= "Variable Categ칩rica", options= text_cols)
    Variable_num= st.sidebar.selectbox(label= "Variable Num칠rica", options= numeric_cols)
    
#GRAPH 4: BARPLOT
#Despliegue de un bar plot, definiendo las variables "X categorias" y "Y num칠ricas" 
    figure4 = px.bar(data_frame=df, x=df[Variable_cat], 
                  y= df[Variable_num], title= str('Features of')+' '+'Passengers')
    figure4.update_xaxes(automargin=True)
    figure4.update_yaxes(automargin=True)
    st.plotly_chart(figure4)
```
